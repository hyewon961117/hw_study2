# 1. 시계열 분석

```python
'''
회귀분석    regression
분류    classification
클러스터링  clustering
시계열분석  time series

자연어처리 - 텍스트분류, 텍스트생성
이미지처리 - 이미지분류, 이미지생성
'''
```

```python
'''
- 시계열 분석(Time Series Analysis) : 주어진 시계열 데이터를 사용하여 기계학습 모형을 만들어서 미래에 일어날 일들을 예측하는 것
- 시계열 데이터 : 시간의 흐름에 따라 관찰된 데이터
- 시계열 분석의 목적 : 과거를 분석하여 미래를 예측하기 위한 것
- 시계열 분석의 가정 : 오늘의 데이터가 내일의 데이터에 영향을 준다

- 회귀 분석과의 차이점:
   변수가 자기상관성(Autocorrelation)을 가지는 것을 전제로 함
   회귀분석과 달리 시계열은 순서가 중요함
   
- 자기상관성: 시간의 변화에 따른 변수간의 상관관계( t 시점의 데이터가 (t-1) 시점, (t-2) 시점, (t-n) 시점의 데이터와 얼마나 관계가 있는가? )
- 정상성(stationarify) : 시간에 따라 확률적인 성분(평균, 분산)이 변하지 않는다는 특징, 평균과 분산이 안정되어 있어서 분석하기 쉬움
- 정상성을 만족하지 않는 경우 분석하기 어려우므로 정상성을 갖도록 전처리를 해야 함(일반적으로 평균이 일정하지 않으면 차분을 취하고 분산이 일정하지 않으면 로그 변환 등을 하게 된다.)

- 안정적인 시계열 : 시간의 추이와 관계없이 평균,분산이 일정한 경우
- 불안정적인 시계열 : 평균이 일정하지 않은 시계열은 차분(difference)을 통해 안정적으로 변환한 후 분석

- 차분(difference) : 현시점 자료에서 전시점 자료를 뺀 값
  1차 차분 : 현시점 자료에서 직전 시점의 자료를 뺀 값
  2차 차분 : 현시점 자료에서 전전 시점의 자료를 뺀 값

- AR모형(자기상관모형, Autocorrelation) : 이전의 값이 이후의 값에 영향을 미치는 모형
  AR(1) : 직전 데이터가 다음 데이터에 영향을 줄 경우
    자기상관함수 ( autocorrelation function : ACF ) : 시간의 진행에 따른 자기 상관의 정도를 나타내는 함수, 자기와 자기 이외의 시차 간의 상관계수를 구하는 함수, yt와 yt-1의 상관관계를 구해서 yt-2 이상의 시차 상관계수를 유추하는 것
        이 수치가 높으면 전 시차의 영향을 계속 받기 때문에 데이터가 독립적이지 않은 것을 의미함 => 불안정한 데이터 => 안정적인 시계열로 변환하는 작업이 필요함
    부분자기상관함수 ( partial autocorrelation function : PACF )
        세 개의 변수 X, Y, Z 가 있을 때 Z 의 효과를 제거한 X와 Y 간의 부분상관관계, yt와 yt-1만의 상관계수를 구하는 것, yt-2 이상의 시차는 고려하지 않음

- MA모형(이동평균, Moving Average) : 시간이 지날수록 평균값이 지속적으로 증가하거나 감소하는 경향
  MA(1) 직전 데이터가 다음 데이터에 영향을 주는 경우

- ARIMA 모형: AR모형과 MA모형을 결합한 모형
  arima(p,d,q)
    p : ar(p) 모형의 p  
    d : 차분차수
    q : ma(q) 모형의 q
'''
```

## 1) 영국왕 수명

```python
#영국왕들의 수명 데이터
import pandas as pd

df = pd.read_csv('./data/time/kings.dat',header=None,names=['age'])
df.head()
```

```python
df.plot()
# 평균과 분산이 일정하지않은 불안정한 상태 
```

```python
# 안정적인 시계열 : 시간의 추이와 관계없이 평균,분산이 일정한 경우
# 불안정적인 시계열 : 차분을 통해 시계열을 안정적으로 변환한 후 분석 진행
# 1025 1030 1015 1030
# 1030-1025=5
# 1015-1030=-15
# 데이터.diff(periods=n) n 차분
# 정상성을 만족시키기 위해 차분을 사용함
```

```python
import matplotlib.pyplot as plt

a=pd.Series([1025,1050,1010,1020,1040])
b=a.diff(periods=1).iloc[1:]
#         차분차수
plt.plot(b)
plt.show()
```

```python
# 정상성(stationary) : 모든 시점에 평균이 일정한 특성
# 시간에 따라 확률적인 성분이 변하지 않는다는 가정, 정상성이 없으면 비시계열 자료로 분류됨
# 차분(difference) : 현시점 자료에서 전시점 자료를 빼는 것
# 평균이 일정하지 않은 시계열은 차분(difference)을 통해 정상화
```

```python
#1차 차분을 통해 데이터를 정상화하는 과정
diff1=df['age'].diff(periods=1).iloc[1:]
diff1.plot()

plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller
import numpy as np

#로그 후 차분한 자료를 adfuller 함수로 안정적인 시계열인지 확인
diff1=np.log(df['age']).diff(periods=1).iloc[1:]
result = adfuller(diff1)
#           안정적인 시계열 데이터인지 검사하는 함수

print(result)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1]) # 유의확률

#p-value가 0.05보다 작으므로 95% 신뢰수준 하에서 유의함(안정적인 시계열 자료임)

# 가설
# 귀무가설 H0 : 불안정한 시계열 데이터
# 대립가설 H1 : 내가 주장하고 싶은 가설, 안정적인 시계열 데이터
```

```python
# 가장 적절한 arima 모델을 추천해주는 함수
# arima: AR모형과 MA모형을 결합한 함수
# arima(p,d,q)
# ar(p) 모형의 p,  
#   자기상관모형(Autocorrelation) : 이전의 값이 이후의 값에 영향을 미치는 모형
#   ar(1) : 직전 데이터가 다음 데이터에 영향을 줄 경우
#   ex) 용수철을 잡아 당기면 원래 길이로 돌아가려는 경향
# ma(q) 모형의 q
#   이동평균(Moving Average) : 시간이 지날수록 평균값이 지속적으로 증가하거나 감소하는 경향
#   ma(1) 직전 데이터가 다음 데이터에 영향을 주는 경우
# d : 차분차수
```

```python
#pip install pmdarima
from pmdarima.arima import auto_arima

model = auto_arima(df)
model

#가장 적절한 모형은 arima(0,1,1)
```

```python
#학습용,검증용 구분
train = df[:30] # 0~29
test = df[30:] # 30~

print(len(df['age']))
print(df['age'].values)
```

```python
#arima(1,2,1)을 수행할 경우  order=c(1,2,1)로 모형을 생성함
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(train, order=(0,1,1)).fit()
model.plot_diagnostics(figsize=(16, 8))

plt.show()

#Standardized residual: 잔차를 시계열로 그린 그래프
#Histogram plus estimated density와 Normal Q-Q : 잔차가 정규성을 만족하는가
#Normal Q-Q plot 데이터의 정규성 가정에 대한 검토, 모집단이 정규성을 따른다면 직선의 형태로 그려짐
#안정적인 시계열 자료이고 정규분포임
```

```python
predictions = model.predict(start=1, end=len(test), dynamic=False) # 한단계 앞선 예측
predictions.index=len(train)+(predictions.index-1)
predictions
```

```python
plt.plot(df,label='Real')
plt.plot(predictions,label='Predict')
plt.legend(loc='upper left',fontsize=8)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error  
mse = mean_squared_error(test, predictions)  # 출력, 모형출력
rmse = np.sqrt(mse)
rmse
```

```python
#1차 차분 데이터
import matplotlib.pyplot as plt

diff1=df['Ozone'].diff(periods=1).iloc[1:]
diff1.plot()
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller
import numpy as np

#로그 후 차분한 자료를 adfuller 함수로 안정적인 시계열인지 확인
diff1=np.log(df['Ozone']).diff(periods=1).iloc[1:]
result=adfuller(diff1)

print(result)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])

#p-value가 0.05보다 작으므로 95% 신뢰수준 하에서 유의함(안정적인 시계열 자료임)
```

```python
plt.plot(df['Ozone'],label='Real')
plt.plot(predictions,label='Predict')
plt.legend(loc='upper left',fontsize=8)
plt.show()
```

```python
import numpy as np

def _error(real, pred):
    return real - pred

def mse(real, pred):
    return np.mean(np.square(_error(real, pred)))

def rmse(real, pred):
    return np.sqrt(mse(real, pred))

def mae(real, pred):
    return np.mean(np.abs(_error(real, pred)))

METRICS = {
    "mse": mse,
    "rmse": rmse,
    "mae": mae,
}

def evaluate(real, pred, metrics=("mse", "rmse", "mae")):
    results = {}
    for name in metrics:
        try:
            results[name] = METRICS[name](real, pred)
        except:
            results[name] = np.nan
    return results
```

```python
evaluate(test['Ozone'].values, predictions.values)
```

# 2. 순환신경망(RNN)

```python
'''
- 순환 신경망(Recurrent neural network, RNN)
유닛간의 연결이 순환적 구조를 갖는 신경망

- 일반적인 신경망
각각의 입출력이 서로 독립적
입력층에서 출력층 한 방향으로만 흐르는 feed forward 신경망

- 기존 신경망과의 차이점
내부의 메모리를 이용해 시퀀스 형태의 입력을 처리할 수 있다.
순서가 있는 정보를 입력 데이터로 사용한다는 점이 특징

- 응용 분야
필기체 인식, 자연어 처리(NL, Natural Language), 음성 인식, 작곡, 동영상 분석, 주가예측 등과 같은 순차적인(sequential) 시계열(time series) 데이터에 적합한 모델

- 순차데이터 - 샘플에 순서가 있는 데이터
  예) 문장
- 타임 스텝 - 순차 데이터를 처리하는 각 단계

good morning => 단어라면 타임 스텝 2, 글자라면 타임 스텝 12
- 순환층 : 순환 구조가 있는 층(은닉층의 출력이 다시 은닉층의 입력으로 사용됨)

현재의 은닉 상태(hidden state)를 계산하기 위해 이전 타임 스텝의 은닉 상태를 사용함

RNN에서는 tahn(하이퍼볼릭 탄젠트, hyperbolic tangent)를 활성화 함수로 많이 사용함
활성화함수 : 순환층 - tanh, 출력층 – sigmoid

학습이 진행될수록 단어들의 연관성을 반영하게 됨

- LSTM(Long short term memory) : RNN 모형의 일종으로 다양한 입력 시퀀스(sequence)를 받아 출력 시퀀스를 만들 수 있는 모형

기본 순환신경망과 달리 2개의 출력이 순환됨
셀 상태(C) : 셀로 순환만 되는 출력
H : 은닉 상태 - 다음 셀로 전달되는 출력

'''
```

## 1) 사인곡선

```python
import numpy as np

a=np.arange(0, 10, 0.5)
print(len(a),a) #0~10까지 0.5 간격으로 값 생성

data = np.sin(a)
print(data)
print(len(data))
```

```python
# dropout 다음 레이어로 넘어갈때 다음 레이어의 몇%의 출력신호를 꺼버림
# return_sequences=True /// False이면 최종 1개만 보내는데 아닌 경우 1개 이상 보냄
```

```python
import pandas as pd

df = pd.read_csv('./data/time/pollution.csv', index_col=0)
df
```

```python
df.boxplot()
```

```python
from sklearn.preprocessing import LabelEncoder
import numpy as np

values = df.values
encoder = LabelEncoder()

# w_dr 변수 정수인코딩
values[:,4] = encoder.fit_transform(values[:,4])
values = values.astype('float32')

np.unique(values[:,4])
```

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled = scaler.fit_transform(values)

pd.DataFrame(scaled).boxplot()
plt.show()
```

```python
#쉬프트
def shift(train):
  timestep = 4

  X = []
  y = []

  for i in range(timestep, len(train)):
    X.append(train[i-timestep:i,:]) # pollution ~ rain
    y.append(train[i,0:1]) # pollution

  return X,y
```

```python
X, y = shift(scaled)

X = np.array(X)
y = np.array(y)

print(X.shape)
print(y.shape)
```

```python
#학습용:검증용 8:2 (4년:1년)
n_train= int(len(X)*.2)+1

X_train, X_test = X[n_train:,] , X[:n_train,]
print(X_train.shape, X_test.shape)

y_train, y_test = y[n_train:,] , y[:n_train,]

print(y_train.shape)
print(y_test.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout,LSTM

model = Sequential()

#return_sequences = False 기본값, 마지막 스텝만 내보내는 옵션
#return_sequences = True, 모든 스텝에 내보내는 옵션(LSTM을 여러층으로 만들 경우 사용)

model.add(LSTM(units = 50, return_sequences = True, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dropout(0.2))
model.add(LSTM(units = 50, return_sequences = True))
model.add(Dropout(0.2))
model.add(LSTM(units = 50))
model.add(Dropout(0.2))
model.add(Dense(units = 1))

model.compile(optimizer = 'adam', loss = 'mse')    

model.summary()
```

```python
model.fit(X_train, y_train, validation_split=0.2, epochs = 10, batch_size = 32)
```

```python
from sklearn.metrics import mean_squared_error

pred = model.predict(X_test)
mse = mean_squared_error(y_test, pred)
rmse = np.sqrt(mse)
rmse
```

```python
d = scaled[:n_train,:]

y_predicted = np.concatenate((pred,d[:n_train,1:]), axis =1)
y_tested = np.concatenate((y_test, d[:n_train,1:]), axis = 1)
```

```python
y_predicted = scaler.inverse_transform(y_predicted)
y_tested = scaler.inverse_transform(y_tested)

y_predicted = y_predicted[:,0:1]
y_tested = y_tested[:,0:1]
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_tested, y_predicted)
rmse = np.sqrt(mse)
rmse
```

```python
from sklearn.metrics import mean_squared_error  

mse = mean_squared_error(y_tested, y_predicted)  

print('RMSE' , np.sqrt(mse))
```

```python
resultsDict = {}
predictionsDict = {}

split_date = '2014-01-01'

df_training = air_pollution.loc[air_pollution.index <= split_date]
df_test = air_pollution.loc[air_pollution.index > split_date]

print(len(df_training))
print(len(df_test))
```

```python
import pmdarima as pm

autoModel = pm.auto_arima(df_training.pollution_today,
                          error_action='ignore', suppress_warnings=True, seasonal=False)

autoModel.fit(df_training.pollution_today)
```

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(df_training.pollution_today, order=(0,0,3)).fit()
```

```python
import numpy as np
from sklearn.metrics import r2_score

def _error(real, pred):
    return real - pred

def mse(real, pred):
    return np.mean(np.square(_error(real, pred)))

def rmse(real, pred):
    return np.sqrt(mse(real, pred))

def mae(real, pred):
    return np.mean(np.abs(_error(real, pred)))

METRICS = {
    "mse": mse,
    "rmse": rmse,
    "mae": mae,
    "r2": r2_score,
}

def evaluate(real, pred, metrics=("mse", "rmse", "mae", "r2")):
    results = {}

    for name in metrics:
        try:
            results[name] = METRICS[name](real, pred)
        except:
            results[name] = np.nan

    return results
```

```python
# pip install tqdm
from tqdm import tqdm
#시간이 오래걸림

index = len(df_training)
yhat = list()

for t in tqdm(range(len(df_test.pollution_today))):
    temp_train = air_pollution[:len(df_training)+t]
    model = ARIMA(temp_train.pollution_today, order=(0, 0, 3))
    model_fit = model.fit()
    predictions = model_fit.predict(
        start=len(temp_train), end=len(temp_train), dynamic=False)
    yhat = yhat + [predictions]

yhat = pd.concat(yhat)

resultsDict['ARIMA'] = evaluate(df_test.pollution_today, yhat.values)
predictionsDict['ARIMA'] = yhat.values
```

```python
def create_time_features(df, target=None):
    
    df['date'] = df.index
    df['hour'] = df['date'].dt.hour
    df['dayofweek'] = df['date'].dt.dayofweek
    df['quarter'] = df['date'].dt.quarter
    df['month'] = df['date'].dt.month
    df['year'] = df['date'].dt.year
    df['dayofyear'] = df['date'].dt.dayofyear
    df['sin_day'] = np.sin(df['dayofyear'])
    df['cos_day'] = np.cos(df['dayofyear'])
    df['dayofmonth'] = df['date'].dt.day
    # df['weekofyear'] = df['date'].dt.weekofyear

    X = df.drop(['date'], axis=1)

    if target:
        y = df[target]
        X = X.drop([target], axis=1)
        return X, y

    return X
```

```python
from sklearn.preprocessing import StandardScaler

X_train_df, y_train = create_time_features(df_training, target='pollution_today')
X_test_df, y_test = create_time_features(df_test, target='pollution_today')

scaler = StandardScaler()
scaler.fit(X_train_df)  

X_train = scaler.transform(X_train_df)
X_test = scaler.transform(X_test_df)

X_train_df = pd.DataFrame(X_train, columns=X_train_df.columns)
X_test_df = pd.DataFrame(X_test, columns=X_test_df.columns)
```

```python
import tensorflow as tf

tf.random.set_seed(0)

BATCH_SIZE = 64
BUFFER_SIZE = 100
WINDOW_LENGTH = 24

#타임스텝 7로 설정
def window_data(X, Y, window=7):
    x = []
    y = []

    for i in range(window-1, len(X)):
        x.append(X[i-window+1:i+1])
        y.append(Y[i])
    return np.array(x), np.array(y)

X_w = np.concatenate((X_train, X_test))
y_w = np.concatenate((y_train, y_test))
X_w, y_w = window_data(X_w, y_w, window=WINDOW_LENGTH)

X_train_w = X_w[:-len(X_test)]
y_train_w = y_w[:-len(X_test)]

X_test_w = X_w[-len(X_test):]
y_test_w = y_w[-len(X_test):]

train_data = tf.data.Dataset.from_tensor_slices((X_train_w, y_train_w))
train_data = train_data.cache().shuffle(BUFFER_SIZE).batch(BATCH_SIZE).repeat()

val_data = tf.data.Dataset.from_tensor_slices((X_test_w, y_test_w))
val_data = val_data.batch(BATCH_SIZE).repeat()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM,Dense

dropout = 0.0

lstm_model = Sequential([
    LSTM(128, input_shape=X_train_w.shape[-2:], dropout=dropout),
    Dense(128),
    Dense(128),
    Dense(1)
])

lstm_model.compile(optimizer='rmsprop', loss='mae')
lstm_model.summary()
```

```python
#steps_per_epoch : 훈련 샘플수 / 배치 사이즈
model_history = lstm_model.fit(train_data, epochs=10,
                                      steps_per_epoch=200,
                                      validation_data=val_data, validation_steps=50)
```

```python
yhat = lstm_model.predict(X_test_w).reshape(1, -1)[0]
resultsDict['LSTM'] = evaluate(y_test, yhat)
predictionsDict['LSTM'] = yhat
```

```python
def bar_metrics(resultsDict):
    df = pd.DataFrame.from_dict(resultsDict)
    colors = plt.rcParams["axes.prop_cycle"].by_key()["color"]
    pallette = plt.cm.get_cmap("tab20c", len(df.columns))
    colors = [pallette(x) for x in range(len(df.columns))]
    color_dict = dict(zip(df.columns, colors))
    fig = plt.figure(figsize=(20, 15))
    fields=['mae','mse','rmse','r2']

    for idx,field in enumerate(fields):
        fig.add_subplot(2, 2, idx+1)
        plt.title(field.upper())
        df.loc[field].sort_values().plot(
            kind="bar",
            colormap="Paired",
            color=[color_dict.get(x, "#333333") for x in df.loc[field].sort_values().index],
        )
        plt.legend()

    plt.tight_layout()
    plt.show()
```

```python
rcParams['xtick.labelsize'] = 15
rcParams['ytick.labelsize'] = 15
bar_metrics(resultsDict)
```
